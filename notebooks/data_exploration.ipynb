```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Analyze Target Variable

```python
#Distribution of Price

plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
sns.histplot(df["price"],kde=True)
plt.title("Price Distribution")

plt.subplot(1,2,2)
sns.boxplot(x=df["price"])
plt.title("Price Box Plot")
plt.tight_layout()
plt.show()
```

```python
# BHK Distribution
df["bhk"].value_counts().plot(kind='bar')
plt.title("BHK Distribution")
```

```python
# Correlation HeatMap
numerical_cols = ['bath', 'balcony', 'price', 'total_sqft_int', 'bhk', 'price_per_sqft']
plt.figure(figsize=(8, 6))
sns.heatmap(df[numerical_cols].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Scatter plots
# prive vs Total Sqft
plt.scatter(df["total_sqft_int"],df["price"],alpha=0.5)
plt.xlabel("Total Sqft")
plt.ylabel("Price ")
plt.title("Price vs Total Sqft")
```
