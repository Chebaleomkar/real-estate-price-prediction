```python
import numpy as np
import pandas as pd 
```

```python
df = pd.read_csv("../data/raw/bangalore house price prediction.csv")
```

```python
df.duplicated().sum()
```

```python
# remove duplicated rows
df.drop_duplicates(inplace=True)

```

```python
df.duplicated().sum() # no duplicate rows now
```

```python
df.isnull().sum() # no null values
```

```python
df.select_dtypes(include=['object']).columns
```

Detect outlier in price

```python

Q1 = df["price_per_sqft"].quantile(0.25)
Q3 = df["price_per_sqft"].quantile(0.75)
IQR = Q3-Q1

lower_bound = Q1 - 1.5 *IQR
upper_bound = Q3 + 1.5 * IQR

print(f"Price per sqft should be between {lower_bound} and {upper_bound}")
```

```python
df.shape[0]
```

Remove Outlier


```python
df = df[(df["price_per_sqft"] >= lower_bound ) & (df["price_per_sqft"] <= upper_bound )]
```

```python
df.shape[0]
```

```python
# Check if sqft per bhk is reasonable (min 300 sqft per bhk)
df = df[df['total_sqft_int'] / df['bhk'] >= 300]
```

```python
df = df[df['bath'] < df['bhk'] + 2]
```

```python
df.to_csv("../data/processed/cleaned_data.csv" , index=False)
print(f"Saved {len(df)} rows after cleaning")
```
