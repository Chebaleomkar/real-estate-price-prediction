```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("../data/processed/cleaned_data.csv")
df.shape
```

```python
df['bath_per_bhk'] = df['bath'] / df['bhk']
df['balcony_per_bhk'] = df['balcony'] / df['bhk']
df['sqft_per_bhk'] = df['total_sqft_int'] / df['bhk']
df['total_amenities'] = df['bath'] + df['balcony']
```

```python
df['log_total_sqft'] = np.log1p(df['total_sqft_int'])
df['log_price'] = np.log1p(df['price'])
```

```python
df['sqft_x_bhk'] = df['total_sqft_int'] * df['bhk']
df['sqft_x_bath'] = df['total_sqft_int'] * df['bath']
df['bhk_x_bath'] = df['bhk'] * df['bath']
```

```python
df['sqft_squared'] = df['total_sqft_int'] ** 2
df['bhk_squared'] = df['bhk'] ** 2
```

```python
df.shape
```

```python
df.to_csv('../data/processed/feature_engineered_data.csv', index=False)
print(f"Saved {len(df)} rows with {df.shape[1]} features")
```
